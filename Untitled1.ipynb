```python
import requests
from bs4 import BeautifulSoup

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.0.0 Safari/537.36"
}

html_text = requests.get('https://wuzzuf.net/search/jobs?a=spbg&q=machine%20learning', headers=headers)
print(html_text.text[:1000])


soup=BeautifulSoup(html_text.text,'lxml')
Ai_eng = soup.find_all('div', class_='css-pkv5bu')
for job in Ai_eng:

        eng_name = job.find('h2', class_='css-m604qf').text.strip()
        eng_location = job.find('span', class_='css-50s5g7').text.strip()
        days_posted_element = job.find('div', class_='css-4c4ojb') or job.find('div', class_='css-do600m')
        days_posted = days_posted_element.text.strip() if days_posted_element else "N/A"


        job_link_element = job.find('a', class_='css-o171kl')
        job_link = job_link_element['href'] if job_link_element else "N/A"


        print(f"Job Title: {eng_name}")
        print(f"Location: {eng_location}")
        print(f"Days Posted: {days_posted}")
        print(f"Job Link: {job_link}")
        print("-" * 20)
        print(html_text.text[:1000])
```

```python
from bs4 import BeautifulSoup

with open('/content/54 machine learning jobs in Egypt – discover job details now on Wuzzuf!.html', 'r', encoding='utf-8') as html_file:
    content = html_file.read()
    soup = BeautifulSoup(content, 'lxml')

    jobs = soup.find_all('div', class_='css-pkv5bu')

    for job in jobs:
        title = job.find('h2', class_='css-m604qf')
        location = job.find('span', class_='css-50s5g7')
        posted = job.find('div', class_='css-4c4ojb') or job.find('div', class_='css-do600m')
        link = job.find('a', class_='css-o171kl')

        print(f"Title: {title.text.strip() if title else 'N/A'}")
        print(f"Location: {location.text.strip() if location else 'N/A'}")
        print(f"Posted: {posted.text.strip() if posted else 'N/A'}")
        print(f"Link: {'https://wuzzuf.net' + link['href'] if link else 'N/A'}")
        print("-" * 30)
```
